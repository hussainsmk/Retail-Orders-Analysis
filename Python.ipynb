```python
#treating 'Not Available' and 'unknown' as null
df = pd.read_csv('orders.csv', na_values = ['Not Available','unknown'])
#checking unique values again
df['Ship Mode'].unique()
```

```python
#changing column names to lower case and replacing spaces with underscores
df.columns=df.columns.str.lower()
```

```python
df.columns
```

```python
df.columns=df.columns.str.replace(' ','_')
```

```python
df.head(5)
```

```python
#creating a new column that will calculate the discount given. This will be done by multiplying list_price with discount_percent
df['discount']=df['list_price']*df['discount_percent']*.01
```

```python
#we will now calculate sale price that will be dervied by subtractng discount from list price
df['sale_price']=df['list_price']-df['discount']
```

```python
df.head(5)
```

```python
#we will now calculate the profit that can be derived by subtracting cost_price from sale_price
df['profit']=df['sale_price']-df['cost_price']
```

```python
df.head(5)
```

```python
#converting order_date into date type
df['order_date']=pd.to_datetime(df['order_date'],format="%Y-%m-%d")
```

```python
#dropping columns list_price, cost_price and discount_percent
df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
```
